## Tenseur Benchmarks

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scienceplots
plt.style.use("science")
```

```python
sizes = np.array([64, 128, 256, 512, 1024, 2048, 4096])
```

```python
ten_openblas = pd.read_csv("./tenseur_openblas.csv", sep = ";")
ten_blis = pd.read_csv("./tenseur_blis.csv", sep = ";")
eigen = pd.read_csv("./eigen.csv", sep = ";")
torch = pd.read_csv("./torch.csv", sep = ";")
```

## Gemm

```python
ten_openblas_gemm = ten_openblas[ten_openblas["name"]=="Gemm"]
ten_blis_gemm = ten_blis[ten_blis["name"]=="Gemm"]
eigen_gemm = eigen[eigen["name"]=="Gemm"]
torch_gemm = torch[torch["name"]=="Gemm"]
fig, ax = plt.subplots()
ax.plot(sizes, ten_openblas_gemm["total"], color = "red", label = "Tenseur openblas")
ax.plot(sizes, ten_blis_gemm["total"], color = "green", label = "Tenseur blis")
ax.plot(sizes, eigen_gemm["total"], color = "blue", label = "Eigen")
ax.plot(sizes, torch_gemm["total"], color = "black", label = "Torch")
ax.set_xlabel(sizes)
ax.legend()
ax.set_title("Matrix multiplication")
plt.savefig("gemm.png")
```

## Chained matrix multiplication

```python
ten_blis_chained = ten_blis[ten_blis["name"]=="Chained"]
ten_openblas_chained = ten_openblas[ten_openblas["name"]=="Chained"]
eigen_chained = eigen[eigen["name"]=="Chained"]
torch_chained = torch[torch["name"]=="Chained"]
fig, ax = plt.subplots()
ax.plot(sizes, ten_openblas_chained["total"], color = "red", label = "Tenseur openblas")
ax.plot(sizes, ten_blis_chained["total"], color = "green", label = "Tenseur blis")
ax.plot(sizes, eigen_chained["total"], color = "blue", label = "Eigen")
ax.plot(sizes, torch_chained["total"], color = "black", label = "Torch")
ax.set_xlabel(sizes)
ax.legend()
ax.set_title("Chained matrix multiplication")
plt.savefig("gemm_chained.png")
```
